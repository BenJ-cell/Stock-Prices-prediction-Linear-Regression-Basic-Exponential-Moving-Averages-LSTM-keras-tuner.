```python
from sklearn.model_selection import train_test_split
# Linear regression Model for stock prediction 
train_x, test_x, train_y, test_y = train_test_split(X.drop(columns='Close'), X['Close'], test_size=0.15 , shuffle=False,random_state = 0)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix, accuracy_score, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
regression = LinearRegression()
regression.fit(train_x, train_y)
print("regression coefficient",regression.coef_)
print("regression intercept",regression.intercept_)
# the coefficient of determination R² 
regression_confidence = regression.score(test_x, test_y)
print("linear regression confidence: ", regression_confidence)
predicted=regression.predict(test_x)
print('Mean Absolute Percentage Error (MAPE):', mean_absolute_percentage_error(test_y, predicted))
print('Mean Absolute Error (MAE):', mean_absolute_error(test_y, predicted))
print('Mean Squared Error (MSE) :', mean_squared_error(test_y, predicted))
print('Root Mean Squared Error (RMSE):', np.sqrt(mean_squared_error(test_y, predicted)))
dfr=pd.DataFrame({'Actual_Price':test_y, 'Predicted_Price':predicted})
dfr.head(10)
```

```python
x2 = dfr.Actual_Price.mean()
y2 = dfr.Predicted_Price.mean()
Accuracy1 = x2/y2*100
print("The accuracy of the model is " , Accuracy1)

plt.scatter(dfr.Actual_Price, dfr.Predicted_Price,  color='Darkblue')
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.show()
```

```python
mean_ma_mape = 0
mean_ema_mape = 0

window_size = 50
rolling_close = pd.DataFrame()

for col in close_df.columns:
  window_var = col + str(window_size) + 'day'
  window_ema_var = window_var+'_EMA'
  # Calculate the 50 days Moving-Average
  rolling_close[window_var] = close_df[col].rolling(window_size).mean()
  # Calculate the 50-day exponentially weighted moving average
  rolling_close[window_ema_var] = close_df[col].ewm(span=window_size, adjust=False).mean()
  # Compute ma mape and ema mape
  mean_ma_mape+=mean_absolute_percentage_error(rolling_close.dropna()[window_var],close_df[col][49:])/5
  mean_ema_mape+=mean_absolute_percentage_error(rolling_close.dropna()[window_ema_var],close_df[col][49:])/5

rolling_close = rolling_close.dropna()
print('mean_ma_mape:',mean_ma_mape)
print('mean_ema_mape:',mean_ema_mape)
rolling_close.head()
```

```python
from sklearn.preprocessing import StandardScaler
# scale our dataset
scaler = StandardScaler()
X_train = X[:-50]
X_test = X[-50:]
scaled_data = scaler.fit_transform(X_train)
X_train = scaled_data
X_test = scaler.transform(X_test)

# scale our dataset
scaler = StandardScaler()
y_train = close_df[:-50]
y_test = close_df[-50:]
scaled_data = scaler.fit_transform(y_train)
y_train = scaled_data
y_test = scaler.transform(y_test)

print('Train shape',X_train.shape,y_train.shape)
print('Test shape',X_test.shape,y_test.shape)
```

```python
import keras_tuner as kt
import tensorflow as tf
from keras.models import Sequential
from tensorflow.keras import regularizers
from keras.layers import Dense, Dropout, LSTM
class MyHyperModel(kt.HyperModel):
  def build(self, hp):
    model = Sequential()
    # hp_units = hp.Int('units', min_value=16, max_value=2048, step=32)
    model.add(LSTM(1712,return_sequences=True,input_shape=(X_train.shape[1], 1), activation="tanh"))
    # model.add(Dropout(rate=hp.Float('dropout_1',min_value=0.0,max_value=0.5,step=0.05)))
    model.add(LSTM(1712,return_sequences=True, activation="tanh"))
    # model.add(Dropout(rate=hp.Float('dropout_2',min_value=0.0,max_value=0.5,step=0.05)))
    model.add(LSTM(1712,return_sequences=True, activation="tanh"))
    # model.add(Dropout(rate=hp.Float('dropout_3',min_value=0.0,max_value=0.5,step=0.05)))
    model.add(Dense(y_test.shape[1], activation='linear'))
    # hp_learning_rate = hp.Choice('learning_rate', values=[1e-3])
    model.compile(loss='mse', optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3))
    return model

tuner = kt.Hyperband(MyHyperModel(), objective='val_loss', max_epochs=50)

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2,patience=20)
EarlyStopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=50)

tuner.search(X_train, y_train, epochs=100, validation_split=0.2, callbacks=[reduce_lr,EarlyStopping])
# Get the optimal hyperparameters
best_hps=tuner.get_best_hyperparameters(num_trials=1)[0]

model = tuner.hypermodel.build(best_hps)
history = model.fit(X_train, y_train, epochs=100, validation_split=0.2, callbacks=[reduce_lr,EarlyStopping])

val_loss_per_epoch = history.history['val_loss']
best_epoch = val_loss_per_epoch.index(max(val_loss_per_epoch)) + 1
print('Best epoch: %d' % (best_epoch,))
```
